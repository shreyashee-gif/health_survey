```python
import polars as pl
```

```python
reverse_coding_items = pl.read_csv("./data/ReverseCodingItems.csv", eol_char="\r")
```

```python
reverse_coding_items
```

```python
health_survey = pl.read_csv("./data/health_survey.csv")
```

```python
health_survey
```

```python
stack = (
    health_survey
    .melt(id_vars="ID")
    .rename({"ID": "Participant", "variable": "Question", "value": "Response"})
)
```

```python
stack
```

```python
reverse_map = reverse_coding_items.select([
    pl.col("Construct").cast(pl.Utf8).alias("Question"),
    pl.col("Needs Reverse Coding?").alias("Needs Reverse"),
])
stack = stack.with_columns(pl.col("Question").cast(pl.Utf8))
stack = stack.join(reverse_map, on = "Question", how = "left")
stack = stack.with_columns(pl.col("Needs Reverse").fill_null("No"))
```

```python
stack
```

```python
responses = {
    "Strongly Disagree": 1,
    "Somewhat Disagree": 2,
    "Neither Agree nor Disagree": 3,
    "Somewhat Agree": 4,
    "Strongly Agree": 5
}
```

```python
stack = stack.with_columns(
    pl.col("Response").replace(responses)
    .cast(pl.Int64).alias("TempRecodeValue")
)
```

```python
stack
```

```python
reverse_scale = {1: 5, 2: 4, 3: 3, 4: 2, 5: 1}
stack = stack.with_columns(
    pl.when(pl.col("Needs Reverse") == "Yes")
    .then(pl.col("TempRecodeValue").replace(reverse_scale))
    .otherwise(pl.col("TempRecodeValue"))
    .alias("TempReverseValue")
)
                      
```

```python
stack
```

```python
stack = stack.with_columns(
    pl.when(pl.col("Needs Reverse") == "Yes")
    .then(pl.col("TempRecodeValue"))
    .otherwise(pl.col("TempRecodeValue"))
    .alias("Recoded Value")
)
```

```python
stack
```

```python
stack = stack.with_columns(
    pl.col("Question").str.split(".").list.first().alias("Question Type")
)
```

```python
stack
```

```python
summary = (
    stack
    .with_columns(
        pl.col("Participant").str.strip_chars().str.replace_all('"', '').cast(pl.Int64)
    )
    .group_by(["Participant", "Question Type"])
    .agg(pl.col("Recoded Value").sum().alias("Recoded Total"))
    .pivot(index="Participant", on="Question Type", values="Recoded Total")
    .select(["Participant", "F1", "F2", "F3", "F4", "F5", "F6"])
    .sort("Participant")
)
```

```python
summary 
```

```python
summary.write_csv("data/health_survey_summary.csv")
```
